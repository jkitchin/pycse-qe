# Using shell commands in Jupyter notebook

The next cell relies on "Jupyter magic" to create an input file for a hydrogen molecule. When you run this cell, it writes the contents out to [example.inp](./example.inp).

```python
%%writefile example.inp
&CONTROL
calculation = 'scf' ,
restart_mode = 'from_scratch' ,
prefix = 'H2' ,
tstress = .true. ,
tprnfor = .true. ,
outdir = '/tmp/$USER' ,
/
&SYSTEM
ibrav = 1 ,
celldm(1) = 10.0 ,
nat = 2 ,
ntyp = 1 ,
ecutwfc = 20 ,
/
&ELECTRONS
conv_thr = 1.0d-8 ,
mixing_mode = 'plain' ,
diagonalization = 'david' ,
/
ATOMIC_SPECIES
H 1.00794 H_ONCV_PBE-1.0.oncvpsp.upf
ATOMIC_POSITIONS bohr
H -0.7 0.0 0.0
H 0.7 0.0 0.0
K_POINTS automatic
1 1 1 0 0 0
```

You can start a cell with ! which means run a shell command. Here we run the calculation and you see the results in the notebook. 

```python
! pw.x < example.inp
```

If the output is very long, you might want to redirect it to a file.

```python
! pw.x < example.inp > example.out
```

Then you can use grep to extract data from the output file. Or, click this link go visit it: [example.out](./example.out).

```python
! grep "!    total energy" example.out
```
